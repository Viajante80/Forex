### Get all instruments from Oanda API
Save to file:
-  instruments.pkl 
- currency_instruments.pkl




```python
import sys
import os

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from oanda_api import OandaApi
import pandas as pd

```

```python
api = OandaApi()
response = api.fetch_instruments()
instruments = response['instruments']
instruments[0].keys()
```

```python
instrument_data = []
for item in instruments:
    new_ob = dict(
        name = item['name'],
        type = item['type'],
        displayName = item['displayName'],
        pipLocation = item['pipLocation'],
        marginRate = item['marginRate']
    )
    instrument_data.append(new_ob)
```

```python
instrument_df = pd.DataFrame.from_dict(instrument_data)
instrument_df
```

```python
instrument_df.to_pickle("../instruments.pkl")
```

```python
currency_instruments_df = instrument_df[instrument_df['type'] == 'CURRENCY']
currency_instruments_df
```

```python
currency_instruments_df.to_pickle("../currency_instruments.pkl")
```
